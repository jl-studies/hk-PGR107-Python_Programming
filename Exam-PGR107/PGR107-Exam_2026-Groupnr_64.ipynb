# Task 1 - Supernacci

```python
#Runner function for starting supernacci without global variables
def runner():
    n = int(input("Enter a number: "))
    supernacci(n)

def supernacci(n):
    #Create a list with the first Supernacci number
    s = [1]

    #Loop from 2 up to n
    #Starts from 2 because s(1) already exists
    for i in range(2, n + 1):
        k = (i - 1) % 3

        #If count is 0, the sum is considered empty, and it will use the previous Supernacci number.
        #In case of 0, it will start at 1.
        if k == 0:
            nextSequence = s[-1]
        else:
            #Select the last k numbers from the list using slicing
            nextSequence = sum(s[-k:])
        #Add the new Supernacci number to the list
        s.append(nextSequence)
    superNacciNumber = s[n-1]

    print("The Supernacci number is:", superNacciNumber)

runner()
```

# Task 4 - Trace on Numpy

```python
import numpy as np
import random as rnd
#Program initializes two random integers to set Matrix size.
#Matrix is generated with two random integers.
#Function trace checks if number of rows and columns are equal with "B.shape"
#The diagonal from matrix is retrieved, sums values and prints sum

#Function for creating the random values, both for the values within the Matrix and the size of the matrix
def randomValues():

    #Generates random integers and sets to variables
    rand1 = rnd.randint(2, 5)
    rand2 = rnd.randint(2, 5)

    #Generates the Matrix with random integer values and sets to variable A
    return np.random.randint(1, 10, size=(rand1, rand2))


#Generates one random Matrix and calculates its trace of its square
def trace():

    #Assigns randomized values to variable A
    A = randomValues()

    #Print statements
    print("Output Matrix:")
    print(A)
    #Print statement for spacing the output matrix and later print statements
    print("")

    #Storage for number of rows and columns from matrix shape
    rows, columns = A.shape

    #Checks if the Matrix is square by comparing the rows to columns
    if rows == columns:
        #Diagonal values of the Matrix
        diagA = np.diag(A)
        #Sum of the diagonal values
        sumA = np.sum(diagA)
        print("The Matrix is square, and the sum is:", sumA)
    else:
        print("The Matrix is not square.")

#Executes program
trace()
```

```python
def averageByCounty():
    #Load the dataset from the CSV.
    df = csvParsing()

    #Group all the data by County and Subject, then calculate the mean value.
    average = (
        df
        .groupby(['County', 'Subject Group'], as_index=False)
        .mean(numeric_only=True)
        .round(1)
    )

    #Replace all the missing values(NaN) with "no data".
    average = average.fillna('No data')


    return average

averageByCounty()
```

The function calculates the grade difference between the two schools for the selected exam or grade column. A positive difference means the first school achieved a higher grade, while a negative difference means the second school performed better.

```python
#Compares the same subject group and exam/grade column between two selected schools.
#Filters the datasat for each school, merges the results, and calculates the grade difference between the two schools.
def schoolVSschool(school1, school2, subject, exam1):
    df = csvParsing()

    #Filters the dataset for the first selected school and subject group.
    filtering1 = df[
        df['School'].str.contains(school1, case=False, na=False) &
        df['Subject Group'].str.contains(subject, case=False, na=False)
        ]

    #Filters the dataset for the second selected school and the same subject group.
    filtering2 = df[
        df['School'].str.contains(school2, case=False, na=False) &
        df['Subject Group'].str.contains(subject, case=False, na=False)
        ]

    #Merge the two filtered dataframes together based on subject group.
    #This makes it so it is possible to compare the two schools on the same subject group.
    mergeSchools = pd.merge(filtering1, filtering2, how='inner', on='Subject Group', suffixes=(f' {school1}', f' {school2}'))

    #difference = combineSchools[exam2] - combineSchools[exam1]

    mergeSchools['Difference'] = mergeSchools[f'{exam1} {school1}'] - mergeSchools[f'{exam1} {school2}']
    mergeSchools['Difference'] = mergeSchools['Difference'].fillna('Not comparable')

    #Renames columns to make the output easier to read.
    mergeSchools = mergeSchools.rename(columns={
        f'School {school1}': 'School 1',
        f'School {school2}': 'School 2',
        f'{exam1} {school1}': 'Grade',
        f'{exam1} {school2}': 'Grade',
    })

    #Replaces remaining missing values with No data
    mergeSchools = mergeSchools.fillna('No data')

    return mergeSchools[
        [
            'Subject Group',
            'School 1',
            'Grade',
            'School 2',
            'Grade',
            'Difference'
        ]
    ]


schoolVSschool('Askim videregående skole', 'Mysen videregående skole', 'Norsk Vg3 studieforberedende utdanningsprogram', '2023/24 Oral - Exam grade')
```
